## Information about the dataset


* https://www.pokemon.com/us/pokedex/
* https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number

This data set includes 721 Pokemon, including their number, name, first and second type, and basic stats: HP, Attack, Defense, Special Attack, Special Defense, and Speed. 

This are the raw attributes that are used for calculating how much damage an attack will do in the games. This dataset is about the pokemon games (NOT pokemon cards or Pokemon Go).

The data as described by Myles O'Neill is:

* #: ID for each pokemon
* Name: Name of each pokemon
* Type 1: Each pokemon has a type, this determines weakness/resistance to attacks or The Primary Type of the Pokemon.
* Type 2: Some pokemon are dual type and have 2 or The Secondary Type of the Pokemon.
* Total: sum of all stats that come after this, a general guide to how strong a pokemon is
* HP: hit points, or health, defines how much damage a pokemon can withstand before fainting
* Attack: the base modifier for normal attacks (eg. Scratch, Punch)
* Defense: the base damage resistance against normal attacks
* SP Atk: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)
* SP Def: the base damage resistance against special attacks
* Speed: determines which pokemon attacks first each round

## Aggregate Statistics (Groupby)

We are going our type 1 column and using that as our indexing column, and filtering our data to get all the defense for the type 1 enries from highest to lowest. So we will see which type has the highest defense. 

```python
## do specific types have good abilities 
poke.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

Now we will see from our type 1 column, which type has the highest attack. 

```python
poke.groupby(['Type 1']).mean().sort_values('Attack', ascending = False)
```

### You can contine and do more filtering for the different columns that affect the  type 1 of the Pokemons. 

You can group by by Type 1 of the Pokemon and simply get the count 'sum' for each column against the type of the various Pokemons.

```python
poke.groupby(['Type 1']).count()
```

This just displays the number of Type 1's for the Pokemons. 

```python
## tidy up the data and get the amounts of each pokemon type
poke['count'] = 1
poke.groupby(['Type 1']).count()['count']
```
